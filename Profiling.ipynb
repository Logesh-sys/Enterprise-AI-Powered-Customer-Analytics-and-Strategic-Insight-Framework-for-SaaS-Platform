```python
customer_intelligence = revenue_df.copy()
```

```python
customer_intelligence['Predicted_Churn'] = churn_df['Predicted_Churn']
customer_intelligence['Churn_Probability'] = churn_df['Churn_Probability']
```

```python
customer_intelligence.shape
```

```python
customer_intelligence[['Customer_ID',
                       'Predicted_Churn',
                       'Churn_Probability',
                       'Predicted_Next_Quarter_Revenue',
                       'Cluster_Label']].head()
```

```python
from sklearn.preprocessing import MinMaxScaler

# Select risk features
risk_features = customer_intelligence[
    ['Churn_Probability',
     'Payment_Delay_Days',
     'Renewal_Risk_Flag',
     'Support_Tickets']
].copy()

# Convert all columns to numeric
for col in risk_features.columns:
    risk_features[col] = pd.to_numeric(risk_features[col], errors='coerce')

# Fill missing values
risk_features = risk_features.fillna(0)

# Scale features
scaler = MinMaxScaler()
risk_scaled = scaler.fit_transform(risk_features)

# Calculate Risk Score
customer_intelligence['Risk_Score'] = (
      risk_scaled[:, 0] * 0.50   # Churn Probability
    + risk_scaled[:, 1] * 0.20   # Payment Delay
    + risk_scaled[:, 2] * 0.20   # Renewal Risk
    + risk_scaled[:, 3] * 0.10   # Support Tickets
) * 100

customer_intelligence['Risk_Score'] = customer_intelligence['Risk_Score'].round(2)

# View results
customer_intelligence[['Customer_ID', 'Risk_Score']].head()
```

```python
customer_intelligence[['Renewal_Risk_Flag',
                       'Payment_Delay_Days',
                       'Support_Tickets']].dtypes
```

```python
# Features for revenue potential
revenue_features = customer_intelligence[
    ['Predicted_Next_Quarter_Revenue',
     'Health_Score',
     'Feature_Adoption_Pct',
     'Renewal_Probability']
].copy()

# Scale to 0-1
scaler = MinMaxScaler()
revenue_scaled = scaler.fit_transform(revenue_features)

# Revenue Potential Score
customer_intelligence['Revenue_Potential_Score'] = (
      revenue_scaled[:,0] * 0.50
    + revenue_scaled[:,1] * 0.20
    + revenue_scaled[:,2] * 0.20
    + revenue_scaled[:,3] * 0.10
) * 100

customer_intelligence['Revenue_Potential_Score'] = (
    customer_intelligence['Revenue_Potential_Score'].round(2)
)
```

```python
customer_intelligence[
    ['Customer_ID',
     'Predicted_Next_Quarter_Revenue',
     'Revenue_Potential_Score']
].head()
```

```python
# Calculate Engagement Score
customer_intelligence['Engagement_Score'] = (
      customer_intelligence['Health_Score'] * 0.40
    + customer_intelligence['Feature_Adoption_Pct'] * 0.40
    + (
        customer_intelligence['Daily_Active_Users']
        / customer_intelligence['Daily_Active_Users'].max()
      ) * 100 * 0.20
).round(2)

# Categorize Engagement Level
customer_intelligence['Engagement_Level'] = np.where(
    customer_intelligence['Engagement_Score'] >= 75,
    'High',
    np.where(
        customer_intelligence['Engagement_Score'] >= 50,
        'Medium',
        'Low'
    )
)
```

```python
customer_intelligence[
    ['Customer_ID',
     'Engagement_Score',
     'Engagement_Level']
].head()
```

```python
customer_profile = customer_intelligence[[
    'Customer_ID',
    'Cluster_Label',
    'Predicted_Churn',
    'Churn_Probability',
    'Predicted_Next_Quarter_Revenue',
    'Risk_Score',
    'Revenue_Potential_Score',
    'Engagement_Score',
    'Engagement_Level'
]]

customer_profile.head()
```

```python
customer_profile.to_csv(
    "Customer_Intelligence_Profile.csv",
    index=False
)

print("Customer Intelligence Profile saved successfully.")
```
