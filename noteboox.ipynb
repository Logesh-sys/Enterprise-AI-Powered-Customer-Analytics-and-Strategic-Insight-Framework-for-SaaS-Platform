```python
transaction_features = transactions.groupby('Customer_ID').agg(
    Total_Transactions=('Transaction_ID', 'count'),

    # Revenue Features
    Total_Revenue_USD=('Net_Revenue_USD', 'sum'),
    Avg_Revenue_USD=('Net_Revenue_USD', 'mean'),
    Total_Billed_USD=('Total_Billed_USD', 'sum'),
    Total_AddOn_Revenue=('Add_On_Revenue_USD', 'sum'),

    # Discount & Tax
    Avg_Discount=('Discount_Pct', 'mean'),
    Avg_Tax=('Tax_Rate_Pct', 'mean'),

    # Payment Behaviour
    Avg_Payment_Delay=('Payment_Delay_Days', 'mean'),
    Max_Payment_Delay=('Payment_Delay_Days', 'max')

).reset_index()
```

```python
late_payment = transactions.groupby('Customer_ID')['Payment_Status'].apply(
    lambda x: (x == 'Late').sum()
).reset_index(name='Late_Payments')

failed_payment = transactions.groupby('Customer_ID')['Payment_Status'].apply(
    lambda x: (x == 'Failed').sum()
).reset_index(name='Failed_Payments')

paid_payment = transactions.groupby('Customer_ID')['Payment_Status'].apply(
    lambda x: (x == 'Paid').sum()
).reset_index(name='Paid_Payments')
```

```python
transaction_type = pd.crosstab(
    transactions['Customer_ID'],
    transactions['Transaction_Type']
).reset_index()
```

```python
#Aggregate Usage

usage_features = usage.groupby('Customer_ID').agg(

    # Usage Activity
    Avg_Daily_Active_Users=('Daily_Active_Users', 'mean'),
    Avg_Monthly_Active_Users=('Monthly_Active_Users', 'mean'),

    # Product Adoption
    Avg_Feature_Adoption=('Feature_Adoption_Pct', 'mean'),

    # API Usage
    Avg_API_Calls=('API_Calls_Monthly', 'mean'),

    # License Usage
    Avg_Licenses_Used=('Licenses_Used', 'mean'),
    Avg_Licenses_Purchased=('Licenses_Purchased', 'mean'),
    Avg_License_Utilization=('License_Utilization_Pct', 'mean'),

    # Session Behaviour
    Avg_Sessions=('Sessions_Total', 'mean'),
    Avg_Session_Min=('Avg_Session_Min', 'mean'),
    Avg_Mobile_Sessions=('Mobile_Sessions_Pct', 'mean'),

    # Platform Usage
    Avg_Integrations=('Integrations_Active', 'mean'),
    Avg_Automations=('Automations_Triggered', 'mean'),
    Avg_Reports=('Reports_Generated', 'mean'),

    # Data Usage
    Avg_Data_Records=('Data_Records_Created', 'mean'),
    Avg_Storage=('Storage_Used_GB', 'mean'),

    # System Health
    Avg_Errors=('Errors_Logged', 'mean'),
    Avg_Uptime=('Uptime_Pct', 'mean')

).reset_index()
```

```python
usage_features.shape
usage_features.head()
```

```python
engagement.columns.tolist()
```

```python
#Aggregate Engagement Features
engagement_features = engagement.groupby('Customer_ID').agg(

    Total_Events=('Engagement_Event_ID', 'count'),

    Avg_Resolution_Days=('Resolution_Days', 'mean'),

    Avg_Score=('Score', 'mean')

).reset_index()
```

```python
event_type = pd.crosstab(
    engagement['Customer_ID'],
    engagement['Event_Type']
).reset_index()
```

```python
channel = pd.crosstab(
    engagement['Customer_ID'],
    engagement['Channel']
).reset_index()
```

```python
severity = pd.crosstab(
    engagement['Customer_ID'],
    engagement['Severity']
).reset_index()
```

```python
resolution = pd.crosstab(
    engagement['Customer_ID'],
    engagement['Resolution_Status']
).reset_index()
```

```python
sentiment = pd.crosstab(
    engagement['Customer_ID'],
    engagement['Sentiment']
).reset_index()
```

```python
followup = engagement.groupby('Customer_ID')['Follow_Up_Required'].apply(
    lambda x: (x == 'Yes').sum()
).reset_index(name='FollowUp_Count')
```

```python
sla = engagement.groupby('Customer_ID')['SLA_Breached'].apply(
    lambda x: (x == 'Yes').sum()
).reset_index(name='SLA_Breach_Count')
```

```python
manager = engagement.groupby('Customer_ID')['Escalated_To_Manager'].apply(
    lambda x: (x == 'Yes').sum()
).reset_index(name='Manager_Escalation_Count')
```

```python
engagement_features = (
    engagement_features
    .merge(event_type, on='Customer_ID', how='left')
    .merge(channel, on='Customer_ID', how='left')
    .merge(severity, on='Customer_ID', how='left')
    .merge(resolution, on='Customer_ID', how='left')
    .merge(sentiment, on='Customer_ID', how='left')
    .merge(followup, on='Customer_ID', how='left')
    .merge(sla, on='Customer_ID', how='left')
    .merge(manager, on='Customer_ID', how='left')
)
```

```python
engagement_features.fillna(0, inplace=True)
```

```python
#Final DF
final_df = customers.copy()

final_df = final_df.merge(
    transaction_features,
    on='Customer_ID',
    how='left'
)

final_df = final_df.merge(
    usage_features,
    on='Customer_ID',
    how='left'
)

final_df = final_df.merge(
    engagement_features,
    on='Customer_ID',
    how='left'
)

```

```python
#Chi-Square Test (Categorical Features)
```

```python
from scipy.stats import chi2_contingency

chi_results = []

for col in cat_cols:

    table = pd.crosstab(X[col], y)

    chi2, p, dof, exp = chi2_contingency(table)

    chi_results.append([col, p])

chi_df = pd.DataFrame(
    chi_results,
    columns=['Feature','P_Value']
)

chi_df.sort_values('P_Value')

```

```python
chi_selected = chi_df[chi_df['P_Value'] < 0.05]

chi_selected
```

```python
#ANOVA (Numerical Features)
```

```python
from scipy.stats import f_oneway

anova_results = []

for col in num_cols:

    churn0 = X[y==0][col]

    churn1 = X[y==1][col]

    stat,p = f_oneway(churn0,churn1)

    anova_results.append([col,p])

anova_df = pd.DataFrame(
    anova_results,
    columns=['Feature','P_Value']
)

anova_df.sort_values('P_Value').head(30)
```

```python
anova_selected = anova_df[anova_df['P_Value']<0.05]

anova_selected
```
